```python
href_list = []
for page_num in tqdm(range(1, 100)):
    url = 'https://kin.naver.com/qna/expertAnswerList.naver?dirId=60221&queryTime=2022-05-31%2000%3A40%3A19&page=' + str(page_num)
    html = requests.get(url)
    soup = bs(html.text, 'html.parser')
    find_tbody = soup.find('tbody', id='au_board_list')
    for i in range(0, 20):
        name = find_tbody.find_all('tr')[i].find('td', class_='questioner').text
        href = find_tbody.find_all('tr')[i].find('td', class_='questioner').find('a')['href']
        href_list.append([name, href])
df = pd.DataFrame(href_list, columns=['변호사이름', '프로필링크'])
df
```
